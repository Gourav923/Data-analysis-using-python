##### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

##### Now read the csv file

```python
df = pd.read_csv('diwalisales.csv',encoding='unicode_escape') #assigning diwalisales.csv to df
#to avoid encoding error, use unicode_escape
```

### Clean the raw data and make it relevant

```python
df.info()
```

```python
#drop/remove/delete the unrelated/blank columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
#cheching the columns Status and unnamed is remove or not
df.info()
#you can also check it by shape function or any other function
```

```python
#checking the null values
df.isnull().sum()
```

```python
#delete the null values
df.dropna(inplace=True)
```

```python
#now checking that the null values delete or not
df.isnull().sum()
```

```python
#change the datatype
df['Amount']=df['Amount'].astype('int')
#astype is used to change the datatype
```

```python
#Now check that the datatype is change or not
df['Amount'].dtypes
```

```python
#all columns names
df.columns
```

```python
#rename column
df.rename(columns={'User_ID':'Cus_ID'},inplace=True)
```

```python
#describe() method returns description of the data in dataframe (i.e. count, mean,std, etc)
df.describe()
```

```python
#use the specified columns
df[['Age','Orders','Amount']].describe()
```

## Exploratory Data Analysis

#### Gender

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=(4,4))
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

###### From above graph we can see the most of the buyers are females and even the purchasing power of the females are stronger then man

### AGE

```python
plt.figure(figsize=(6,6))
ax=sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
plt.figure(figsize=(4,4))
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

###### From above graphs we can see the most of the buyers are of age group b/w 26-35yrs female

### State

```python
# total numn=ber of orders from top 10 states
plt.figure(figsize=(16,5))
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#total amount/sales from top 10 states
plt.figure(figsize=(16,5))
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x='State',y='Amount',data=sales_state)
```

##### From the above graphs we can see that unexpectedly most of the orders from UP,Maharashtra,and Karnataka respectivly but total sales/amount is from UP,Karnataka and then Maharashtra

### Marital Status

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='Marital_Status', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(5,5))
marital=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Marital_Status',y='Amount',data=marital, hue='Gender')
```

##### From above graphs we can see thatthe most of the buyers are married(woman) and they have high purchasing power

### Occupation

```python
plt.figure(figsize=(18,6))
ax=sns.countplot(x='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(18,5))
occu=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Occupation',y='Amount',data=occu)
```

##### From the above graphs we can see that most of the buyes are working in IT, Healthcare and Aviation sector

### Product Category

```python
plt.figure(figsize=(22,10))
ax=sns.countplot(x='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(26,9))
PC=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Product_Category',y='Amount',data=PC)
```

##### From above graphs we can see the most of the sold products are from foods and electronic category

```python
plt.figure(figsize=(26,9))
PC=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(x='Product_ID',y='Orders',data=PC)
```

```python
#top 10 most sold products(same thing as above)
fig1, ax1=plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
